```python
# Start with loading all necessary libraries
import numpy as np
import nltk
import pandas as pd
from os import path
from PIL import Image
from wordcloud import WordCloud, ImageColorGenerator
from PIL import Image

import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
```

```python
from urllib.request import Request, urlopen
from bs4 import BeautifulSoup

link = Request('http://g1.globo.com/politica/noticia/2019/08/11/mais-da-metade-da-camara-ja-teve-desconto-no-salario-por-ausencia-valor-total-chega-a-r-12-milhao.ghtml', headers={'User-Agent': 'Mozilla/5.0'})
pagina = urlopen(link).read().decode('utf-8', 'ignore')
soup = BeautifulSoup(pagina, "lxml")
textos = soup.find_all('p', attrs={'class':'content-text__container'})
len(textos)
```

```python
paragrafos = []
for tag in textos:
    paragrafos.append(tag.text)
    



```

```python
df = pd.read_csv('bonoro.csv')
df
```

```python
for text in df["Noticia"]:
    paragrafos.append(text)
```

```python
file = open('bolsonaroNoticias.txt', 'w')

    
texto = ''.join(paragrafos)
texto.replace('\n', '')
file.write(texto)
file.close
```

```python
texto
```

```python
stopwords = nltk.corpus.stopwords.words('portuguese')
wordcloud = WordCloud(width=1920, height=1080, background_color='white', min_font_size = 10, stopwords=stopwords).generate(texto)
wordcloud.to_file("bonoroCloud.png")
# Display the generated image:
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```

```python
bonoro_mask = np.array(Image.open("peixe.png"))
bonoro_mask
```

```python
wc = WordCloud(width=1820, height=2770, background_color='white', min_font_size = 10, stopwords=stopwords, mask=bonoro_mask, max_words=2000).generate(texto)
wc.to_file("bonoroCloudMask.png")
# Display the generated image:
plt.imshow(wc, interpolation='bilinear')
plt.axis("off")
plt.show()
```
